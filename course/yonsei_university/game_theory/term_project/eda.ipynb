### Single Sample Feature Values

- 변수들이 어떠한 관계가 있어보임 (이렇게 스무스할 수가 없음)

```python
normal = train_set.data.query('target == 1')
abnormal = train_set.data.query('target == -1')
normal_idx = np.random.randint(0, len(normal))
abnormal_idx = np.random.randint(0, len(abnormal))

normal_X = normal.iloc[normal_idx, :-1]
abnormal_X = abnormal.iloc[abnormal_idx, :-1]
```

```python
plot_config = {
    'title': f'Normal_{normal_idx}',
    'xticks': xticks,
    'ylabel': 'Sensor Value',
}
fig, (ax1, ax2, ax3) = plt.subplots(
    3, 1,
    figsize=(15, 10),
    gridspec_kw={'height_ratios': [1, 1, 2]},
)
suptitle = 'Compare normal and abnormal sensor value'
fig.suptitle(suptitle)
eda.plot_feature_value(normal_X, plot_config, ax=ax1)

plot_config['title'] = f'Abnormal_{abnormal_idx}'
eda.plot_feature_value(abnormal_X,
                       plot_config, color='#F97306', ax=ax2)

plot_config['title'] = 'Normal/Abnormal'
eda.plot_feature_value(normal_X, plot_config, color='lightgrey',
                       label=f'Normal_{normal_idx}', ax=ax3)
eda.plot_feature_value(abnormal_X, plot_config,
                       label=f'Abnormal_{abnormal_idx}', ax=ax3)
plt.tight_layout()
```
